```python

import matplotlib.pyplot as plt
import numpy as np

# Here I am changing the style of my graphs with grayscale
plt.style.use('grayscale')


# Here I am defining the mandelbrot function, with N being our number of iterations, min/max being the boundaries
def mandelbrot(N, x_min, x_max, y_min, y_max):

    # matrix_nxn is an (N x N) empty array
    matrix_nxn = np.empty([N, N], float)

    # This bit creates a small step in the x, and y axis
    delta_x = abs((x_max - x_min)) / N
    delta_y = abs((y_max - y_min)) / N


    # These for loops are for generating the actual image, going through loop i1, and then
    # staying in i2 up to N. Afterwards, we go back to i1 + 1, and stay in i2 up to N, then i1 + 2, etc...
    for i1 in range(N):
        for i2 in range(N):

            # The x value location on the x-axis
            x_loop = x_min + (delta_x * i1)
            # The y value location on the y-axis
            y_loop = y_min + (delta_y * i2)

            # c is our complex value in f(z) = z**2 + c, it will vary in y_loop up to N, while x_loop
            # is constant, then x_loop increments by 1, and y_loop varies up to N again, etc...
            c = complex(x_loop, y_loop)
            # Initial z value
            z = complex(0.0, 0.0)


            # Dummy variable i, for the while loop
            i = 0

            # The while loop iterates through z, while being inside of the for loops, here we check our
            # conditions that must be satisfied
            while(abs(z) < 2 and i <= 200):
                z = z*z + c
                i += 1

            # This indicates that the value is not in the mandelbrot set
            if(abs(z) >= 2.0):
                matrix_nxn[i2, i1] = 1
            # This is in the set
            else:
                matrix_nxn[i2, i1] = 0

    return matrix_nxn
```

```python
plt.imshow(mandelbrot(1000, -2, 2, -2, 2), extent=[-2, 2, -2, 2])
plt.xlabel("Real Numbers")
plt.ylabel("Imaginary Numbers")
plt.title("Mandelbrot Set")
plt.minorticks_on()
plt.show()
```
